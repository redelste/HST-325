```python
import pandas as pd
import datetime

import numpy as np

df1 = pd.read_csv(open('/home/ryebyread/dev/HST-325/HST35 - Final Project/data.csv'), encoding='utf-8', engine='c')

```

```python
df1 = df1[['text','created_at','is_retweet']]
```

```python
df1 = df1.dropna(axis=0, how='any')
```

```python
def fixDate(row):
    try:
        rmTime = row['created_at'].split(' ')[0]
        rmDash = rmTime.split('-')
        rmDash.reverse()
        if ''.join(rmDash) == '':
            row['created_at']= None
        else:
            row['created_at']= ''.join(rmDash)
        return row
    
    except:
        row['created_at'] = None
        return row
#lambda row goes through every single row
df1 = df1.apply(lambda row: fixDate(row), axis=1)
```

```python

df1['randNumCol'] = np.random.randint(1, 9000, df1.shape[0])

df1Real = df1.dropna(axis=0, how='any')
```

```python
dateDups = df1Real.pivot_table(index=['created_at'], aggfunc='size')
```

```python
df1Real.sort_values(by=['created_at'])
```

```python
df1Real['ID'] = df1Real.apply( lambda row: dateDups.loc[row.loc['created_at']], axis = 1)
```

```python
# remove all non-integers in "created_at" column
df1Real.apply(lambda row: row if row['created_at'].isdigit() else None, axis = 1 )
```

```python
# remove duplicated dates while maintaining weights. 
df1Real = df1Real.drop_duplicates(['created_at'])
```

```python
df1Real.to_csv('data1.csv')
```
